# 이태원더버거 해시태그 인스타그램 크롤링

```python
baseUrl = "https://www.instagram.com/explore/tags/"
plusUrl = input('검색할 태그를 입력하세요 : ')
url = baseUrl + quote_plus(plusUrl)

driver = webdriver.Chrome(
    executable_path="./webdriver/chromedriver.exe"
)
driver.get(url)
```

```python
SCROLL_PAUSE_TIME = 5.0
reallink = []

while True:
    pageString = driver.page_source
    bsObj = BeautifulSoup(pageString, 'lxml')

    for link1 in bsObj.find_all(name='div', attrs={"class":"Nnq7C weEfm"}):
        for i in range(3):
            title = link1.select('a')[i]
            real = title.attrs['href']
            reallink.append(real)

    last_height = driver.execute_script('return document.body.scrollHeight')
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")

    if new_height == last_height:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(SCROLL_PAUSE_TIME)
        new_height = driver.execute_script("return document.body.scrollHeight")

        if new_height == last_height:
            break
        else:
            last_height = new_height
            continue
```

```python
from urllib.request import HTTPError

num_of_data = len(reallink)

print('총 {0}개의 데이터를 수집합니다.'.format(num_of_data))
csvtext = []

for i in tqdm(range(num_of_data)):

    csvtext.append([])
    req = Request("https://www.instagram.com/p"+reallink[i], headers={'User-Agent': 'Mozila/5.0'})
    try:
        webpage = urlopen(req).read()
    except HTTPError as e:
        print(e)
    else:
        soup = BeautifulSoup(webpage, 'lxml', from_encoding='utf-8')
        soup1 = soup.find('meta', attrs={'property':"og:description"})

        reallink1 = soup1['content']
        reallink1 = reallink1[reallink1.find("@") + 1:reallink1.find(")")]
        reallink1 = reallink1[:20]

        if reallink1 == '':
            reallink1 = "Null"
        csvtext[i].append(reallink1)

        for reallink2 in soup.find_all('meta', attrs={'property':"instapp:hashtags"}):
            hashtags = reallink2['content'].rstrip(',')
            csvtext[i].append(hashtags)

        # csv로 저장

        data = pd.DataFrame(csvtext)
        data.to_csv('ItawonTheBurger.txt', encoding='utf-8')

driver.close()
```

# 명사, 키워드 추출

```python
from konlpy.tag import Twitter
from collections import Counter

def tag_counting(file, nlpy, kinds, encoding, hashTags):
    
    f = open(file, "r", encoding = encoding)
    lines = f.read()
    
    #nlpy = Twitter()
    nouns = nlpy.nouns(lines)

    count = Counter(nouns) 

    tag_count = []
    tags = []
#     hashTags = pd.DataFrame()

    for n, c in count.most_common(100):
        dics = {'tag': n, 'count': c}

        if len(dics['tag']) >= 2 and len(tags) <= 49:
            tag_count.append(dics)
            tags.append(dics['tag'])

    print('['+kinds+']')
    for tag in tag_count:
        hashTags[tag['tag']] = tag['count']
        print(" {:<14}".format(tag['tag']), end='\t')
        print("{}".format(tag['count']))

    print("명사 총  {}개". format(len(tags)))
    return tags
```

```python
nlpy = Twitter()
burger_frequency = pd.DataFrame(index = ['fre'])
burger_frequency = burger_frequency.fillna(0)
hashTags = tag_counting(file='ItawonTheBurger.txt', 
                        nlpy = nlpy, kinds = 'hashTag_Twitter',
                        encoding='utf-8', 
                        hashTags = burger_frequency)
```

# 사용자의 인스타그램 해시태그가 '이태원더버거'의 키워드를 언급한 횟수

```python
def frequency_count(file, nlpy, kinds, encoding):
    
    f = open(file, "r", encoding = encoding)
    lines = f.read()
    
    #nlpy = Twitter()
    nouns = nlpy.nouns(lines)

    count = Counter(nouns) 
    
    for n in count:
        if len(n) >= 2 and len(n) <= 49:
            if n in frequency.columns:
                frequency[n]+=int(count[n])
```

```python
nlpy = Twitter()
frequency_count(file='user.txt', nlpy = nlpy, kinds = 'User_Twitter', encoding='utf-8')
frequency
```

# 오토인코더 활용 (빈도 -> 벡터)

```python
import numpy as np

data = np.array([], dtype = int)

for x in frequency.columns:
    val = frequency.loc['fre',x]
#     print(val)
    data = np.append(data,np.array([val]))
    
data = data.reshape(1,1,50)
data
```

```python
from keras.layers import Input, Dense
from keras.models import Model

input_vec = Input(shape=(1,50), name = 'input',kernel_initializer='he_normal')
encoded = Dense(40, activation='relu', name = 'encoder_1')(input_vec)
encoded = Dense(30, activation='relu', name = 'encoder_2')(encoded)
encoded = Dense(20, activation='relu', name = 'encoder_3')(encoded)

decoded = Dense(20, activation='relu', name = 'decoder_1')(encoded)
decoded = Dense(30, activation='relu', name = 'decoder_2')(decoded)
decoded = Dense(40, activation='relu',name = 'decoder_3')(decoded)
decoded = Dense(50, activation='sigmoid',name = 'decoder_4')(decoded)

autoencoder = Model(input_vec, decoded)
autoencoder.compile(optimizer='adadelta', loss='MSE')

autoencoder.summary()
```

```python
x_train = data

autoencoder.fit(x_train, x_train,
                epochs=100,
                batch_size=1)
                #validation_data=(x_test, x_test))
```

```python
print(data)
```

```python
predicted = np.squeeze(autoencoder.predict(data, batch_size=1))

print(predicted)
```
